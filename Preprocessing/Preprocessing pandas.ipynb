## Import the module

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Read the Data

```python
data = pd.read_csv("Data.csv")
data.head()
```

## Take copy of data

```python
df = data.copy()
```

```python
df
```

```python
df.describe()
```

```python
df.isnull().any()
```

```python
df["Age"].fillna(value=df.Age.mean(),axis=0,inplace=True)
```

```python
df
```

```python
df["Salary"].fillna(value=df.Salary.mean(),axis=0,inplace=True)
```

```python
df
```

```python
df.isnull().any()
```

```python
df.head()
```

```python
df.Purchased = pd.factorize(df.Purchased)[0]
```

```python
df.head()
```

```python
dum = pd.get_dummies(df.Country,drop_first=True)
```

```python
df = pd.concat([dum,df],axis = 1)
```

```python
df
```

```python
df.drop(columns=["Country"],axis = 0, inplace = True)
```
